```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# create data for each tasks
# permuted_mnist contains 10 tasks, each consists of 1 classes
x_tasks = [1, 2, 3, 4, 5]
baselines = [51.92, 51.83, 51.36, 51.56, 98.79]  # 61.09
gmm = [64.05, 63.17, 62.42, 62.72, 99.41] # 70.35
pr =  [95.61, 95.48, 91.97, 94.21, 99.91] # 95.43
gmm_pr = [97.43, 96.45, 95.38, 96.11, 99.87]  # 97.04

# plot lines
plt.title("Task-Given + SplitCifar10")
plt.plot(x_tasks, baselines, label = "baselines", marker='o')
plt.plot(x_tasks, gmm, label = "gmm", marker='o')
plt.plot(x_tasks, pr, label = "pr", marker='o')
plt.plot(x_tasks, gmm_pr, label = "gmm + pr", marker='o')
plt.xlabel("Task")
plt.ylabel("Task Accuracy")
plt.legend()
plt.show()
```
