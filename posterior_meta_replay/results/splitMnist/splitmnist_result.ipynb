```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data for each tasks
# split_mnist contains 5 tasks, each consists of 2 classes
x_tasks = [1, 2, 3, 4, 5]
baselines = [99.95, 94.97, 96.03, 96.24, 95.06]  # 96.45
gmm = [99.95, 99.36, 99.73, 99.84, 99.42]  # 99.66
pr = [99.95, 99.41, 99.75, 99.86, 99.63]  # 99.72
gmm_pr = [99.95, 99.45, 99.77, 99.87, 99.66]  # 99.74


# plot lines
plt.title("Task-Given + SplitMnist")
plt.plot(x_tasks, baselines, label = "baselines", marker='o')
plt.plot(x_tasks, gmm, label = "gmm", marker='o')
plt.plot(x_tasks, pr, label = "pr", marker='o')
plt.plot(x_tasks, gmm_pr, label = "gmm + pr", marker='o')
plt.xlabel("Task")
plt.ylabel("Task Accuracy")
plt.legend()
plt.show()
```
